```python
from time import localtime
import datetime

activities ={
    8: "Sleeping", 9: "Cmmuting", 17: "Working", 18: "Commuting", 20: "Eating", 22: "Resting"
}

time_now = localtime()
hour = time_now.tm_hour
print(hour)

for activities_time in sorted(activities.keys()):
  if (hour < activities_time):
    print (activities[activities_time])
    break
  else:
    print("Unkown, AFK or Sleeping!")
```
